### 데이터 생성 및 plot

```python
# 데이터 생성
df = pd.DataFrame(np.random.randn(100,4), # 100개 4개 컬럼
 columns='A B C D'.split())
print(df.shape)
df.head()
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],
                    'y':[10,20,30,40,60],
                    'z':[5,4,3,2,1]})
df3
```
