```python
# No.1
while True:
    temperature = input("현재온도: ")
    
    try:
        if(float(temperature) >= 27):
            print("반바지를 입으세요.")
            break
        else:
            print("긴바지를 입으세요.")
            break

    except:
        print("숫자만 입력하세요.")
```

```python
# No.2
while True:
    num = input("정수입력: ")
    
    try:
        if(int(num)%2 == 0):
            print("짝수")
            break
        else:
            print("홀수")
            break
    
    except:
        print("정수를 입력하세요.")
```

```python
# No.3
while True:
    score = input("점수입력: ")
    
    try:
        score = float(score)
        if(0 > score or 100 < score):
            print("0에서 100사이의 숫자를 입력하세요.")
            
        elif(95 <= score):
            print("그랜드마스터")
            break
        elif(90 <= score):
            print("커널마스터")
            break
        elif(80 <= score):
            print("익스퍼트")
            break
        elif(70 <= score):
            print("초심자")
            break
            
        else:
            print("해당하지 않음")
            break
    
    except:
        print("숫자만 입력하세요.")
```

```python
# No.5
in_str = input("문자입력: ")
out_str = ""
for i in in_str:
    try:
        int(i)
    except:
        out_str += i
print(out_str)
```

```python
# No.6
import random

while True:
    RPS = ["가위", "바위", "보"]
    my_RPS = input("가위바위보: ")
    com_RPS = RPS[random.randrange(0,3)]
    
    if(my_RPS == "가위"):
        if(com_RPS == "가위"):
            print("무승부")
        elif(com_RPS == "바위"):
            print("패")
            break
        elif(com_RPS == "보"):
            print("승")
            break
    elif(my_RPS == "바위"):
        if(com_RPS == "가위"):
            print("승")
            break
        elif(com_RPS == "바위"):
            print("무승부")
        elif(com_RPS == "보"):
            print("패")
            break
    if(my_RPS == "보"):
        if(com_RPS == "가위"):
            print("패")
            break
        elif(com_RPS == "바위"):
            print("승")
            break
        elif(com_RPS == "보"):
            print("무승부")
    
    else:
        print("다시 입력하세요.")
    
```

```python
# No.7
import random

cnt = 2
num_chance = 3
com_selnum = random.randrange(0,11)

while True:
    my_num = input("숫자 맞추기: ")
    
    try:
        my_num = int(my_num)
        if(cnt > num_chance):
            print("실패! 정답은 %d" % com_selnum)
            break
        else:
            if(0 > my_num or 10 < my_num):
                print("0에서 10사이의 숫자를 입력하세요.")
            elif(com_selnum == my_num):
                print("맞아요.")
                break
            elif(com_selnum > my_num):
                print("더 큰 값이예요. 다시 맞춰봐요.")
                cnt += 1
            elif(com_selnum < my_num):
                print("더 작은 값이예요. 다시 맞춰봐요.")
                cnt += 1

    except:
        print("정수만 입력하세요.")
```
