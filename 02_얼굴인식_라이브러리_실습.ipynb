박나래, 기안84, 성훈, 시언 4명의 얼굴위치를 표시해보자!

````python
# 이미지 읽어오기
image = cv.read('nahon.jpeg')


# 얼굴 위치
locations = [
  [280, 300, 350, 240],   # 박나래
  [220, 540, 280, 480],   # 기안84
  [220, 830, 280, 770],   # 성훈
  [220, 1060, 310, 1000]  # 시언 
]


# for 반복문 출력
for k in locations:
  # 박스 그려주세요.
  cv.box(image, k)


# 이미지 출력
cv.show(image)
````
